```python
# RFMiD Retina HastalÄ±k Tespiti - Model EÄŸitim Notebook'u

# ðŸ“¦ 1. Gerekli KÃ¼tÃ¼phaneler
import pandas as pd
import os
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

# ðŸ“ 2. Dosya YollarÄ±
base_dir = r"D:/Yapay zeka eÄŸitme/Erhan Hoca Proje/RFMiD/data"
train_img_dir = os.path.join(base_dir, "train")
val_img_dir = os.path.join(base_dir, "validation")
test_img_dir = os.path.join(base_dir, "test")

train_csv = os.path.join(base_dir, "RFMiD_Training_Labels.csv")
val_csv = os.path.join(base_dir, "RFMiD_Validation_Labels.csv")
test_csv = os.path.join(base_dir, "RFMiD_Testing_Labels.csv")

# âœ… 3. EÄŸitim Etiketlerini GÃ¼venli Åžekilde YÃ¼kle
if not os.path.exists(train_csv):
    raise FileNotFoundError(f"Etiket dosyasÄ± bulunamadÄ±: {train_csv}\nLÃ¼tfen dosya yolunun doÄŸru olduÄŸundan emin olun.")

df_train = pd.read_csv(train_csv)

print("CSV sÃ¼tunlarÄ±:", df_train.columns.tolist())

# Kolon adÄ± farklÄ± olabilir, esnek kontrol
if "Image No" in df_train.columns:
    image_col = "Image No"
elif "Image_Name" in df_train.columns:
    image_col = "Image_Name"
elif "image" in df_train.columns:
    image_col = "image"
else:
    raise KeyError("CSV dosyasÄ±nda 'Image No' veya benzeri bir sÃ¼tun bulunamadÄ±. SÃ¼tun isimleri: " + ", ".join(df_train.columns))

print("EÄŸitim etiketi Ã¶rnekleri:")
print(df_train.head())

# ðŸ”— 4. GÃ¶rsel Yolunu Ekle (PNG veya JPG olabilir)
def find_image_path(img_name, folder):
    # Ã–ncelikle .png uzantÄ±sÄ± ile dene
    png_path = os.path.join(folder, img_name.replace(".jpg", ".png"))
    if os.path.exists(png_path):
        return png_path
    # Yoksa .jpg olarak deneyelim
    jpg_path = os.path.join(folder, img_name)
    if os.path.exists(jpg_path):
        return jpg_path
    # Dosya bulunamadÄ±
    return None

df_train["image_path"] = df_train[image_col].apply(lambda x: find_image_path(x, train_img_dir))

# Eksik gÃ¶rsel olup olmadÄ±ÄŸÄ±nÄ± kontrol et
missing_images = df_train[df_train["image_path"].isnull()]
if not missing_images.empty:
    raise FileNotFoundError(f"AÅŸaÄŸÄ±daki gÃ¶rseller bulunamadÄ±:\n{missing_images[image_col].tolist()}\nLÃ¼tfen 'train' klasÃ¶rÃ¼ndeki dosyalarÄ± ve uzantÄ±larÄ±nÄ± kontrol edin.")

# ðŸ–¼ï¸ 5. Ã–rnek GÃ¶rsel GÃ¶ster
sample_path = df_train["image_path"].iloc[0]
img = Image.open(sample_path)

plt.imshow(img)
plt.title(f"GÃ¶rsel: {df_train[image_col].iloc[0]}")
plt.axis("off")
plt.show()

# ðŸ”Ž 6. Etiket SÃ¼tunlarÄ±nÄ± Belirle
# Image sÃ¼tunu ve image_path hariÃ§ kalanlar etiket kolonlarÄ±
excluded_cols = [image_col, "image_path"]
label_columns = [col for col in df_train.columns if col not in excluded_cols]

print("Etiket sÃ¼tunlarÄ±:", label_columns)

# GÃ¶rsel baÅŸÄ±na kaÃ§ etiket var?
df_train["label_count"] = df_train[label_columns].sum(axis=1)
print("Ortalama etiket sayÄ±sÄ±:", df_train["label_count"].mean())

# ðŸ“Š 7. Etiket DaÄŸÄ±lÄ±mÄ±
label_counts = df_train[label_columns].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
label_counts.plot(kind="bar")
plt.title("Toplam GÃ¶rsel SayÄ±sÄ± (Her HastalÄ±k)")
plt.xlabel("HastalÄ±k")
plt.ylabel("Adet")
plt.grid(True)
plt.tight_layout()
plt.show()
```
